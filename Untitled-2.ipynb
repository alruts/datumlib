 # Full Image Processing Workflow (Datumlib Version)

 Image deblurring pipeline using PSF simulation + Wiener deconvolution

```python
import numpy as np
from dataclasses import dataclass
from scipy.signal import fftconvolve
from skimage import data
from skimage.util import img_as_float
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from itertools import product

from datumlib import Datum, collect, Pipeline
import datumlib as dl

rng = np.random.default_rng(42)
```

### Add a custom Datum container

```python
@dataclass(frozen=True)
class Image2D(Datum[np.ndarray]):
    def plot(self, title=None):
        import matplotlib.pyplot as plt

        plt.figure(figsize=(3, 3))
        plt.imshow(self.data, cmap="gray")
        plt.axis("off")
        plt.tight_layout()
        plt.show()


# helper to display and plot an Image2D datum
def disp_and_plot(x: Image2D):
    x.plot()
    dl.display_datum(x)
```

```python
clean_image = img_as_float(data.camera())

clean = Image2D(clean_image, tags={"type": "clean_reference"})
disp_and_plot(clean)
```

### Generate PSFs

```python
def generate_psf(size=31, sigma=3.0, kind="gaussian"):
    ax = np.arange(-(size // 2), size // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)

    if kind == "gaussian":
        psf = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))

    elif kind == "motion":
        psf = np.zeros((size, size))
        psf[size // 2, :] = 1.0

    elif kind == "defocus":
        radius = size // 3
        psf = ((xx**2 + yy**2) <= radius**2).astype(float)

    else:
        raise ValueError("Unknown PSF type")

    psf /= psf.sum()
    return psf


psf_types = ["gaussian", "motion", "defocus"]
noise_levels = [0.1, 0.01]

psf_data = []

for kind, noise in product(psf_types, noise_levels):
    psf = generate_psf(kind=kind)
    psf_data.append(Image2D(psf, tags={"psf_type": kind, "noise_std": noise}))

psf_collection = collect(*psf_data, tags={"dataset": "synthetic_psfs"})
psf_collection.map_void(disp_and_plot)
```

```python
def simulate_measurement(psf: Image2D):
    try:
        blurred = fftconvolve(clean.data, psf.data, mode="same")

        noisy = blurred + psf.tags.get("noise_std", 0.01) * rng.standard_normal(
            blurred.shape
        )

        return noisy

    except Exception as e:
        print(f"Simulation error: {e}")
        return None


measurement_collection = psf_collection.map_data(simulate_measurement)
```

```python
measurement_collection.map_void(disp_and_plot)
```

```python
def preprocess(d: Image2D) -> Image2D:
    new_data = np.clip(d.data, 0, 1)
    return d.with_data(new_data)


def wiener_deconvolution(d: Image2D, reg_lambda=1e-3) -> Image2D:

    try:
        psf_type = d.tags["psf_type"]
        psf = next(
            p for p in psf_collection.valid_entries if p.tags["psf_type"] == psf_type
        )

        H = np.fft.fft2(psf.data, s=d.data.shape)
        Y = np.fft.fft2(d.data)

        H_conj = np.conj(H)
        X_est = (H_conj * Y) / (np.abs(H) ** 2 + reg_lambda)

        x_est = np.real(np.fft.ifft2(X_est))

        return d.with_data(x_est).add_tags({"method": "wiener", "lambda": reg_lambda})

    except Exception as e:
        print(f"Wiener error for {d.tags}: {e}")
        return None


def postprocess(d: Image2D) -> Image2D:
    return d.with_data(np.clip(d.data, 0, 1))
```

```python
estimation_pipe = Pipeline(
    {
        "preprocess": preprocess,
        "wiener": wiener_deconvolution,
        "postprocess": postprocess,
    }
)

estimated_collection = estimation_pipe(
    measurement_collection, progress_bar=True, parallel=True
)
```

### Adding metrics

```python
def mse(a: Image2D, b: Image2D):
    return float(np.mean(np.square(a.data - b.data)))


def evaluate(est: Image2D):
    return {
        "MSE": mse(est, clean),
        "PSNR": psnr(clean.data, est.data),
        "SSIM": ssim(clean.data, est.data, data_range=1.0),
    }


estimated_collection = estimated_collection.map_tags(evaluate)
```

```python
estimated_collection.map_void(disp_and_plot)
```

### Easily extract metadata and plot

```python
import matplotlib.pyplot as plt

psf_types = dl.collection_util.get_tags(estimated_collection, "psf_type")
noise_levels = dl.collection_util.get_tags(estimated_collection, "noise_std")

labels = [f"{p}, {n}" for p, n in zip(psf_types, noise_levels)]

for metric in ["PSNR", "MSE", "SSIM"]:
    values = dl.collection_util.get_tags(estimated_collection, metric)

    plt.figure()
    plt.bar(labels, values)
    plt.title(metric)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

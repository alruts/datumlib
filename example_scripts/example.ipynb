```python
import datumlib as dl
import numpy as np
import datetime
```

```python
# generate some random signals
class_pool = ["A", "B", "C"]
operator_pool = ["John", "Bob", "Alice"]
sample_rate = 44100
xs = [
    dl.datum(
        np.random.normal(0, 1, 44100),
        tags={
            "class": np.random.choice(class_pool),
            "operator": np.random.choice(operator_pool),
            "sample_rate": 44100,
        },
    )
    for _ in range(5)
]

collection = dl.collect(
    *xs,
    tags={
        "date": datetime.date.today(),
        "time": datetime.datetime.now().strftime("%H:%M:%S"),
    },
)
dl.display_collection(collection)
```

```python
from datumlib import collection_util as cu
from datumlib import datum_util as du
# some data processing steps


def truncate(x: np.ndarray, sample_rate: int) -> np.ndarray:
    trunc_sec = 1
    return x[: trunc_sec * sample_rate]


truncate_mapped = cu.cmap(du.over_data(truncate))

collection = truncate_mapped(collection)
dl.display_collection(collection)
```

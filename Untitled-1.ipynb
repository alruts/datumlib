# Wiener Filtering Example Using datumlib
## Realistic System Identification / Room IR Recovery

Workflow:
   1. Generate synthetic room impulse responses
   2. Generate white noise excitation signal
   3. Simulate noisy measurements
   4. Recover impulse response via Wiener deconvolution
   5. Evaluate estimation error


```python
import numpy as np

from datumlib import collect, Datum
import datumlib as dl
from datumlib import Pipeline

from dataclasses import dataclass, field
from itertools import product
```

```python
@dataclass(frozen=True)
class TimeSeries(Datum):
    sample_rate: float = field(kw_only=True)

    @classmethod
    def create(cls, data: np.ndarray, sample_rate: float, *, tags={}):
        return cls(data, sample_rate=sample_rate, tags=tags)

    @property
    def n_samples(self) -> int:
        return len(self.data)

    @property
    def duration(self) -> float:
        return self.n_samples / self.sample_rate

    @property
    def is_valid(self) -> bool:
        return len(self.data.shape) == 1

    @property
    def time(self):
        return np.linspace(0, self.duration, self.n_samples)

    def plot(self):
        import matplotlib.pyplot as plt

        plt.plot(self.time, self.data)
        plt.show()
```

```python
rng = np.random.default_rng(42)
sample_rate = 32_000


def generate_rir(length=sample_rate * 3, rt60=0.6, noise=1e-4):
    t = np.arange(length) / sample_rate
    decay = np.exp(-6.91 * t / rt60)  # RT60-consistent decay
    rir = decay * rng.standard_normal(length)
    rir[0] = 1.0  # direct path
    rir += noise * rng.standard_normal(length)
    return rir


sources = {
    "S1": (0.0, 0.0, 0.0),
    "S2": (0.5, 0.5, 0.5),
}

receivers = {
    "R1": (0.1, 0.2, 0.3),
    "R2": (0.6, 0.7, 0.5),
}

signals = []

for (src_id, src_loc), (rec_id, rec_loc) in product(sources.items(), receivers.items()):
    rt60 = rng.uniform(0.4, 1.0)
    rir = generate_rir(rt60=rt60)

    signals.append(
        TimeSeries.create(
            rir,
            sample_rate=sample_rate,
            tags={
                "source_id": src_id,
                "receiver_id": rec_id,
            },
        )
    )

ir_collection = collect(*signals, tags={"dataset": "synthetic_rooms"})
dl.display_collection(ir_collection)


for x in ir_collection.valid_entries:
    x.plot()
```

```python
def generate_excitation(duration=3.0):
    n = int(duration * sample_rate)
    excitation = rng.standard_normal(n)

    return TimeSeries.create(
        excitation,
        sample_rate=sample_rate,
    )


excitation = generate_excitation()
dl.display_datum(excitation)
```

```python
def simulate_measurement(excitation: TimeSeries, true_ir: TimeSeries, noise_std=0.01):
    y = np.convolve(excitation.data, true_ir.data)[: excitation.n_samples]
    y += noise_std * rng.standard_normal(len(y))

    return TimeSeries.create(
        y,
        sample_rate=excitation.sample_rate,
        tags={
            "source_id": true_ir.tags["source_id"],
            "receiver_id": true_ir.tags["receiver_id"],
            "noise_std": noise_std,
        },
    )


measurements = []
for ir in ir_collection.valid_entries:
    measurements.append(simulate_measurement(excitation, ir))

measurement_collection = collect(*measurements)
dl.display_collection(measurement_collection)

for x in measurement_collection.valid_entries:
    x.plot()
```

```python
# Wiener Deconvolution


def wiener_deconvolution(
    excitation: TimeSeries,
    measurement: TimeSeries,
    reg_lambda: float,
):
    n = excitation.n_samples

    X = np.fft.rfft(excitation.data, n=n)
    Y = np.fft.rfft(measurement.data, n=n)

    H_est = (Y * np.conj(X)) / (np.abs(X) ** 2 + reg_lambda)
    h_est = np.fft.irfft(H_est, n=n)

    return measurement.with_data(h_est).add_tags(
        tags={
            "method": "wiener",
            "lambda": reg_lambda,
        }
    )
```

```python
def truncate_ir(d: TimeSeries, duration=1.0) -> TimeSeries:
    new_data = d.data[: int(d.sample_rate * duration)]
    return d.with_data(new_data)


def hp_filter(d: TimeSeries, cutoff=40.0) -> TimeSeries:
    from scipy.signal import sosfiltfilt, butter

    sos = butter(4, cutoff, btype="high", output="sos", fs=d.sample_rate)
    new_data = sosfiltfilt(sos, d.data)
    return d.with_data(new_data)


preprocess = Pipeline(
    {"truncate": truncate_ir, "hp_filter": hp_filter},
)

# validate that labels match
tag_check = dl.collection_util.compare_tags(
    ir_collection, measurement_collection, keys=["source_id", "receiver_id"]
)

# sort by source_id and reciever id
measurement_collection = (
    preprocess.run(measurement_collection, progress_bar=True)
    .sort_by("source_id")
    .sort_by("receiver_id")
)
```
